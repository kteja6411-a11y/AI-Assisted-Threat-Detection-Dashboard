```python
import pandas as pd

data = {
    "Name": ["Rahul", "Priya", "Sneha", "Arjun", "Krishna"],
    "Marks": [88, 93, 81, 96, 89],
    "City": ["Vizag", "Hyderabad", "Chennai", "Bangalore", "Vizag"]
}

df = pd.DataFrame(data)

print("\nStudent Data")
print(df)
```

```python
print("\nFirst Five Records")
print(df.head())

print("\nInformation")
print(df.info())

print("\nStatistics")
print(df.describe())

print("\nMissing Values")
print(df.isnull().sum())
```

```python
print("\nStudents Scoring Above 85")

high_marks = df[df["Marks"] > 85]

print(high_marks)
```

```python
print("\nSorted by Marks")

sorted_df = df.sort_values(by="Marks", ascending=False)

print(sorted_df)
```

```python
print("\nTop 3 Students")

top3 = df.nlargest(3, "Marks")

print(top3)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))

plt.bar(df["Name"], df["Marks"])

plt.title("Student Marks")

plt.xlabel("Students")

plt.ylabel("Marks")

plt.show()
```

```python
plt.figure(figsize=(6,6))

plt.pie(df["Marks"],
        labels=df["Name"],
        autopct="%1.1f%%")

plt.title("Marks Distribution")

plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.plot(df["Name"],
         df["Marks"],
         marker="o",
         linewidth=2)

plt.title("Student Performance")

plt.xlabel("Students")

plt.ylabel("Marks")

plt.grid(True)

plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.hist(df["Marks"], bins=5)

plt.title("Marks Histogram")

plt.xlabel("Marks")

plt.ylabel("Frequency")

plt.show()
```

```python
plt.figure(figsize=(5,5))

plt.boxplot(df["Marks"])

plt.title("Box Plot of Marks")

plt.show()
```
